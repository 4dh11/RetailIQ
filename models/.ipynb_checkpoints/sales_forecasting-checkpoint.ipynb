# 📈 Sales Forecasting Model — Facebook Prophet
**Project:** Intelligent E-Commerce Data Pipeline and Decision Support System

### What this notebook does:
1. Loads and prepares monthly sales data from your pipeline
2. Trains a **Facebook Prophet** time-series model
3. Evaluates accuracy using MAE, RMSE, and MAPE
4. Forecasts the **next 6 months** of revenue
5. Detects **seasonal trends** and **peak months** automatically
6. Exports all charts + a forecast CSV for the Flask dashboard

**Prerequisite:** `data_exploration.ipynb` must have been run first.
- Required file: `../data/processed_retail_data.csv`

---
## 📦 Step 1 — Install & Import Libraries

```python
# Run this once if prophet is not installed
# !pip install prophet
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")

from prophet import Prophet
from prophet.diagnostics import cross_validation, performance_metrics
from prophet.plot import plot_cross_validation_metric

from sklearn.metrics import mean_absolute_error, mean_squared_error

import joblib

print("✅ All libraries imported successfully")
```

---
## 📂 Step 2 — Load & Prepare Data

```python
df = pd.read_csv("../data/processed_retail_data.csv", parse_dates=["InvoiceDate"])

print(f"Loaded: {df.shape[0]:,} rows")
print(f"Date range: {df['InvoiceDate'].min().date()} → {df['InvoiceDate'].max().date()}")
df.head(3)
```

```python
# Aggregate to monthly revenue
df["YearMonth"] = df["InvoiceDate"].dt.to_period("M")

monthly = (
    df.groupby("YearMonth")["Sales"]
    .sum()
    .reset_index()
)
monthly["YearMonth"] = monthly["YearMonth"].dt.to_timestamp()
monthly.columns = ["ds", "y"]   # Prophet requires exactly these column names

# Sort chronologically
monthly = monthly.sort_values("ds").reset_index(drop=True)

print(f"Monthly data points: {len(monthly)}")
print(f"Period: {monthly['ds'].min().strftime('%b %Y')} → {monthly['ds'].max().strftime('%b %Y')}")
monthly
```

```python
# Plot raw monthly sales to understand the pattern before modelling
plt.figure(figsize=(13, 5))
plt.plot(monthly["ds"], monthly["y"] / 1e6, marker="o",
         color="#3498db", linewidth=2, markersize=6)
plt.fill_between(monthly["ds"], monthly["y"] / 1e6,
                 alpha=0.15, color="#3498db")

plt.title("Monthly Revenue — Historical Data (2009–2011)",
          fontsize=14, fontweight="bold")
plt.xlabel("Month")
plt.ylabel("Revenue (£ Millions)")
plt.xticks(rotation=30)
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter("%b %Y"))
plt.grid(axis="y", alpha=0.3)
plt.tight_layout()
plt.savefig("../data/raw_monthly_sales.png", dpi=150, bbox_inches="tight")
plt.show()
print("✅ Saved raw_monthly_sales.png")
```

---
## ✂️ Step 3 — Train / Test Split

We hold out the **last 3 months** as a test set to measure how well the model predicts unseen data before using it for real future forecasts.

```python
TEST_MONTHS = 3

train = monthly.iloc[:-TEST_MONTHS].copy()
test  = monthly.iloc[-TEST_MONTHS:].copy()

print(f"Training set : {len(train)} months  "
      f"({train['ds'].min().strftime('%b %Y')} → {train['ds'].max().strftime('%b %Y')})")
print(f"Test set     : {len(test)} months   "
      f"({test['ds'].min().strftime('%b %Y')} → {test['ds'].max().strftime('%b %Y')})")
```

---
## 🤖 Step 4 — Train Prophet Model

**Why Prophet?**
- Built by Facebook/Meta for business time-series
- Automatically handles yearly + weekly seasonality
- Robust to missing data and outliers
- Gives uncertainty intervals out of the box (great for dashboards)
- No manual feature engineering needed

```python
model = Prophet(
    yearly_seasonality  = True,
    weekly_seasonality  = False,   # monthly data — weekly not relevant
    daily_seasonality   = False,
    seasonality_mode    = "multiplicative",  # better for retail (peaks scale with revenue level)
    changepoint_prior_scale = 0.1,           # controls trend flexibility
    seasonality_prior_scale = 10,            # controls seasonality strength
    interval_width      = 0.95               # 95% confidence interval
)

# Add monthly seasonality explicitly (since our data is monthly)
model.add_seasonality(
    name   = "monthly",
    period = 30.5,
    fourier_order = 5
)

model.fit(train)
print("✅ Prophet model trained successfully")
```

---
## 📊 Step 5 — Evaluate on Test Set

```python
# Predict on test period
future_test = model.make_future_dataframe(periods=TEST_MONTHS, freq="MS")
forecast_all = model.predict(future_test)

# Extract test predictions
test_forecast = forecast_all.tail(TEST_MONTHS)[["ds", "yhat", "yhat_lower", "yhat_upper"]]
test_actual   = test[["ds", "y"]].reset_index(drop=True)
eval_df = test_actual.merge(test_forecast, on="ds")

# Metrics
mae  = mean_absolute_error(eval_df["y"], eval_df["yhat"])
rmse = np.sqrt(mean_squared_error(eval_df["y"], eval_df["yhat"]))
mape = (np.abs((eval_df["y"] - eval_df["yhat"]) / eval_df["y"]).mean()) * 100

print("=" * 45)
print("       MODEL EVALUATION — TEST SET")
print("=" * 45)
print(f"  MAE   : £{mae:>12,.2f}")
print(f"  RMSE  : £{rmse:>12,.2f}")
print(f"  MAPE  : {mape:>10.2f}%")
print("=" * 45)
print("\nPredicted vs Actual:")
for _, row in eval_df.iterrows():
    print(f"  {row['ds'].strftime('%b %Y')}  "
          f"Actual: £{row['y']:>10,.0f}  "
          f"Predicted: £{row['yhat']:>10,.0f}  "
          f"Error: {abs(row['y']-row['yhat'])/row['y']*100:.1f}%")
```

```python
fig, ax = plt.subplots(figsize=(13, 5))

# Training actuals
ax.plot(train["ds"], train["y"] / 1e6, color="#3498db",
        linewidth=2, marker="o", markersize=5, label="Training Actual")

# Test actuals
ax.plot(test["ds"], test["y"] / 1e6, color="#2ecc71",
        linewidth=2.5, marker="o", markersize=8, label="Test Actual")

# Test predictions
ax.plot(eval_df["ds"], eval_df["yhat"] / 1e6, color="#e74c3c",
        linewidth=2.5, marker="s", markersize=8, linestyle="--", label="Test Predicted")

# Confidence interval on test
ax.fill_between(eval_df["ds"],
                eval_df["yhat_lower"] / 1e6,
                eval_df["yhat_upper"] / 1e6,
                alpha=0.2, color="#e74c3c", label="95% Confidence Interval")

ax.axvline(test["ds"].min(), color="gray", linestyle=":",
           linewidth=1.5, label="Train/Test Split")

ax.set_title("Sales Forecast — Model Evaluation (Train vs Test)",
             fontsize=14, fontweight="bold")
ax.set_xlabel("Month")
ax.set_ylabel("Revenue (£ Millions)")
ax.xaxis.set_major_formatter(mdates.DateFormatter("%b %Y"))
plt.xticks(rotation=30)
ax.legend()
ax.grid(axis="y", alpha=0.3)

# Annotate MAPE on chart
ax.text(0.02, 0.95, f"MAPE: {mape:.1f}%",
        transform=ax.transAxes, fontsize=11,
        verticalalignment="top",
        bbox=dict(boxstyle="round", facecolor="white", alpha=0.8))

plt.tight_layout()
plt.savefig("../data/forecast_evaluation.png", dpi=150, bbox_inches="tight")
plt.show()
print("✅ Saved forecast_evaluation.png")
```

---
## 🔮 Step 6 — Retrain on Full Data & Forecast Next 6 Months

Now that we've validated accuracy, we retrain on **all available data** to make the most informed forecast possible.

```python
FORECAST_MONTHS = 6

# Retrain on full dataset
final_model = Prophet(
    yearly_seasonality      = True,
    weekly_seasonality      = False,
    daily_seasonality       = False,
    seasonality_mode        = "multiplicative",
    changepoint_prior_scale = 0.1,
    seasonality_prior_scale = 10,
    interval_width          = 0.95
)
final_model.add_seasonality(name="monthly", period=30.5, fourier_order=5)
final_model.fit(monthly)

# Create future dates
future = final_model.make_future_dataframe(periods=FORECAST_MONTHS, freq="MS")
forecast = final_model.predict(future)

# Split into historical fitted + future forecast
historical_forecast = forecast[forecast["ds"] <= monthly["ds"].max()]
future_forecast     = forecast[forecast["ds"] >  monthly["ds"].max()]

print(f"✅ Forecast generated for {FORECAST_MONTHS} months ahead")
print(f"\nNext {FORECAST_MONTHS} months forecast:")
for _, row in future_forecast[["ds", "yhat", "yhat_lower", "yhat_upper"]].iterrows():
    print(f"  {row['ds'].strftime('%b %Y')}  "
          f"Predicted: £{row['yhat']:>10,.0f}  "
          f"Range: [£{row['yhat_lower']:>10,.0f} – £{row['yhat_upper']:>10,.0f}]")
```

```python
fig, ax = plt.subplots(figsize=(14, 6))

# Historical actual
ax.plot(monthly["ds"], monthly["y"] / 1e6,
        color="#3498db", linewidth=2, marker="o",
        markersize=5, label="Actual Revenue", zorder=3)

# Historical fitted
ax.plot(historical_forecast["ds"], historical_forecast["yhat"] / 1e6,
        color="#95a5a6", linewidth=1.5, linestyle="--",
        label="Model Fit", zorder=2)

# Future forecast
ax.plot(future_forecast["ds"], future_forecast["yhat"] / 1e6,
        color="#e74c3c", linewidth=2.5, marker="D",
        markersize=8, label="Forecast (Next 6 Months)", zorder=3)

# Forecast confidence interval
ax.fill_between(future_forecast["ds"],
                future_forecast["yhat_lower"] / 1e6,
                future_forecast["yhat_upper"] / 1e6,
                alpha=0.2, color="#e74c3c", label="95% Confidence Interval")

# Historical CI
ax.fill_between(historical_forecast["ds"],
                historical_forecast["yhat_lower"] / 1e6,
                historical_forecast["yhat_upper"] / 1e6,
                alpha=0.1, color="#3498db")

# Divider line
ax.axvline(monthly["ds"].max(), color="black", linestyle=":",
           linewidth=1.5, label="Forecast Start")
ax.text(monthly["ds"].max(), ax.get_ylim()[1] * 0.95,
        " Forecast →", fontsize=10, color="black")

ax.set_title("Sales Revenue Forecast — Next 6 Months",
             fontsize=14, fontweight="bold")
ax.set_xlabel("Month")
ax.set_ylabel("Revenue (£ Millions)")
ax.xaxis.set_major_formatter(mdates.DateFormatter("%b %Y"))
plt.xticks(rotation=30)
ax.legend(loc="upper left")
ax.grid(axis="y", alpha=0.3)

plt.tight_layout()
plt.savefig("../data/sales_forecast_6months.png", dpi=150, bbox_inches="tight")
plt.show()
print("✅ Saved sales_forecast_6months.png")
```

---
## 🔍 Step 7 — Trend & Seasonality Decomposition

This is what makes Prophet powerful — it separates your revenue into:
- **Trend** — the long-term growth direction
- **Seasonality** — repeating annual patterns (Q4 peaks, Jan dips etc.)

```python
fig, axes = plt.subplots(3, 1, figsize=(13, 10))

# Trend
axes[0].plot(forecast["ds"], forecast["trend"] / 1e6,
             color="#2c3e50", linewidth=2)
axes[0].fill_between(forecast["ds"],
                     forecast["trend_lower"] / 1e6,
                     forecast["trend_upper"] / 1e6,
                     alpha=0.15, color="#2c3e50")
axes[0].set_title("Long-Term Revenue Trend", fontsize=12, fontweight="bold")
axes[0].set_ylabel("Revenue (£M)")
axes[0].grid(axis="y", alpha=0.3)
axes[0].xaxis.set_major_formatter(mdates.DateFormatter("%b %Y"))

# Yearly seasonality
yearly_cols = [c for c in forecast.columns if "yearly" in c]
if yearly_cols:
    axes[1].plot(forecast["ds"], forecast["yearly"],
                 color="#8e44ad", linewidth=2)
    axes[1].axhline(0, color="gray", linestyle="--", linewidth=0.8)
    axes[1].set_title("Yearly Seasonality Effect", fontsize=12, fontweight="bold")
    axes[1].set_ylabel("Seasonal Effect")
    axes[1].grid(axis="y", alpha=0.3)
    axes[1].xaxis.set_major_formatter(mdates.DateFormatter("%b %Y"))

# Monthly seasonality
monthly_cols = [c for c in forecast.columns if "monthly" in c and "lower" not in c and "upper" not in c]
if monthly_cols:
    axes[2].bar(forecast["ds"], forecast["monthly"],
                color=[
                    "#e74c3c" if v > 0 else "#3498db"
                    for v in forecast["monthly"]
                ],
                width=20, alpha=0.8)
    axes[2].axhline(0, color="gray", linestyle="--", linewidth=0.8)
    axes[2].set_title("Monthly Seasonality Effect", fontsize=12, fontweight="bold")
    axes[2].set_ylabel("Seasonal Effect")
    axes[2].grid(axis="y", alpha=0.3)
    axes[2].xaxis.set_major_formatter(mdates.DateFormatter("%b %Y"))

plt.setp([a.get_xticklabels() for a in axes], rotation=25)
plt.suptitle("Revenue Decomposition — Trend & Seasonality",
             fontsize=14, fontweight="bold", y=1.01)
plt.tight_layout()
plt.savefig("../data/forecast_decomposition.png", dpi=150, bbox_inches="tight")
plt.show()
print("✅ Saved forecast_decomposition.png")
```

---
## 📅 Step 8 — Peak Month & Business Insights

```python
# Average revenue by calendar month (seasonality pattern)
monthly["month_num"] = monthly["ds"].dt.month
monthly["month_name"] = monthly["ds"].dt.strftime("%b")

monthly_avg = monthly.groupby(["month_num", "month_name"])["y"].mean().reset_index()
monthly_avg = monthly_avg.sort_values("month_num")

peak_month   = monthly_avg.loc[monthly_avg["y"].idxmax(), "month_name"]
lowest_month = monthly_avg.loc[monthly_avg["y"].idxmin(), "month_name"]

print(f"📈 Peak revenue month   : {peak_month}")
print(f"📉 Lowest revenue month : {lowest_month}")

fig, ax = plt.subplots(figsize=(11, 5))
colors = [
    "#e74c3c" if m == peak_month
    else "#3498db" if m == lowest_month
    else "#bdc3c7"
    for m in monthly_avg["month_name"]
]
bars = ax.bar(monthly_avg["month_name"], monthly_avg["y"] / 1e6,
              color=colors, edgecolor="white", linewidth=1.2)

for bar in bars:
    ax.text(bar.get_x() + bar.get_width() / 2,
            bar.get_height() + 0.01,
            f"£{bar.get_height():.2f}M",
            ha="center", va="bottom", fontsize=8.5)

ax.set_title("Average Monthly Revenue by Calendar Month",
             fontsize=13, fontweight="bold")
ax.set_xlabel("Month")
ax.set_ylabel("Avg Revenue (£ Millions)")

# Legend
from matplotlib.patches import Patch
legend_elements = [
    Patch(facecolor="#e74c3c", label=f"Peak Month ({peak_month})"),
    Patch(facecolor="#3498db", label=f"Lowest Month ({lowest_month})"),
    Patch(facecolor="#bdc3c7", label="Other Months")
]
ax.legend(handles=legend_elements)
ax.grid(axis="y", alpha=0.3)

plt.tight_layout()
plt.savefig("../data/seasonal_pattern.png", dpi=150, bbox_inches="tight")
plt.show()
print("✅ Saved seasonal_pattern.png")
```

```python
# Year-over-Year growth analysis
monthly["year"] = monthly["ds"].dt.year
yoy = monthly.groupby("year")["y"].sum().reset_index()
yoy.columns = ["Year", "Total_Revenue"]
yoy["YoY_Growth_%"] = yoy["Total_Revenue"].pct_change() * 100

print("Year-over-Year Revenue Summary:")
print("=" * 45)
for _, row in yoy.iterrows():
    growth_str = f"{row['YoY_Growth_%']:+.1f}%" if not pd.isna(row["YoY_Growth_%"]) else "baseline"
    print(f"  {int(row['Year'])}  Revenue: £{row['Total_Revenue']:>12,.0f}   Growth: {growth_str}")
print("=" * 45)
```

---
## ✅ Step 9 — Cross-Validation (Model Robustness Check)

```python
# Prophet's built-in cross-validation
# initial: train on first 15 months, horizon: predict 3 months ahead
df_cv = cross_validation(
    final_model,
    initial = "456 days",   # ~15 months
    period  = "30 days",    # retrain every month
    horizon = "90 days",    # evaluate 3 months ahead
    parallel = None
)

df_perf = performance_metrics(df_cv)

print("Cross-Validation Performance Metrics:")
print(df_perf[["horizon", "mae", "rmse", "mape"]].to_string(index=False))
```

```python
fig = plot_cross_validation_metric(df_cv, metric="mape")
fig.set_size_inches(11, 4)
fig.axes[0].set_title("MAPE Across Forecast Horizons (Cross-Validation)",
                      fontsize=13, fontweight="bold")
plt.tight_layout()
plt.savefig("../data/cross_validation_mape.png", dpi=150, bbox_inches="tight")
plt.show()
print("✅ Saved cross_validation_mape.png")
```

---
## 💾 Step 10 — Save All Outputs

```python
# 1. Future forecast CSV — main dashboard data
forecast_export = future_forecast[["ds", "yhat", "yhat_lower", "yhat_upper"]].copy()
forecast_export.columns = ["Month", "Predicted_Revenue", "Lower_Bound", "Upper_Bound"]
forecast_export["Month"] = forecast_export["Month"].dt.strftime("%Y-%m")
forecast_export = forecast_export.round(2)
forecast_export.to_csv("../data/sales_forecast.csv", index=False)
print(f"✅ Saved sales_forecast.csv")

# 2. Full forecast including history — for the full line chart
full_export = forecast[["ds", "yhat", "yhat_lower", "yhat_upper", "trend"]].copy()
full_export.columns = ["Month", "Predicted_Revenue", "Lower_Bound", "Upper_Bound", "Trend"]
full_export["Month"] = full_export["Month"].dt.strftime("%Y-%m")
full_export = full_export.round(2)
full_export.to_csv("../data/sales_forecast_full.csv", index=False)
print(f"✅ Saved sales_forecast_full.csv")

# 3. Monthly historical + averages — for seasonal pattern chart
monthly_avg.to_csv("../data/monthly_seasonality.csv", index=False)
print(f"✅ Saved monthly_seasonality.csv")

# 4. YoY summary — for stats cards on dashboard
yoy.to_csv("../data/yoy_growth.csv", index=False)
print(f"✅ Saved yoy_growth.csv")

# 5. Model evaluation metrics — for dashboard display
metrics = pd.DataFrame([{
    "MAE":  round(mae, 2),
    "RMSE": round(rmse, 2),
    "MAPE": round(mape, 2)
}])
metrics.to_csv("../data/forecast_metrics.csv", index=False)
print(f"✅ Saved forecast_metrics.csv")

# 6. Trained model
joblib.dump(final_model, "../data/prophet_model.pkl")
print(f"✅ Saved prophet_model.pkl")

print("\n🎉 All outputs saved. Ready for Flask dashboard!")
```
